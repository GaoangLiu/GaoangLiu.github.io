A naive implementation of Decision Tree

```python
import numpy as np
```

First, we define our entropy function
$$E = - \sum p(X) \cdot \log_2(p(X))$$
where $p(X) = \frac{\#x}{n}$, $\#x$ is the number of occurrences and $n$ total number of samples

ID3 decides as the best feature with maximum information gain:
$$IG = H(s) - E(D|f_i)$$

Gini index/split
$$\text{Gini} = 1 - \sum_{i=1}^n (p_i)^2 $$

* Favors larger paritions 
* Uses squared proportion of classes
* Perfectly classified, i.e., all items belong to the same class , Gini Index would be zero.
* Evenly distributed would be $1 - 1 / N$


```python
def entropy(y):
    '''Function to calculate the entropy of input array y
    E.g., entropy([0, 1]) = 1.0
    '''
    frequencies = Counter(y).values()
    total = sum(frequencies)
    return - sum(map(lambda f: (f/total) * np.log2(f/total), frequencies))
            
    '''For binary classification 
    hist = np.bincount(y)
    ps = hist / len(y)
    return -np.sum([p * np.log2(p) for p in ps if p > 0]) '''
    

def gini(y):
    '''Gini index'''
    cnt = Counter(y).values()
    if len(cnt) < 1: return 0
    total = sum(cnt)
    return 1 - sum(map(lambda c: c ** 2, cnt))  / (total ** 2)

y = [0,0,0,1,1,1,0,0,1,1]
entropy(y)
gini([1, 1])
```

```python
from collections import Counter

class Node:
    def __init__(self, feature=None, threshold=None, left=None, right=None, *, value=None):
        self.feature = feature
        self.threshold = threshold
        self.left = left
        self.right = right
        self.value = value 
    
    def is_leaf_node(self) -> bool:
        '''A node if leaf node if and only if it has a value
        '''
        return self.value is not None

    
class DecisionTree:
    def __init__(self, min_samples_split=2, max_depth=10, n_features=None, criterion='entropy'):
        '''random factor: n_features are randomly selected, this is also the
        reason why it is called Random Forest if we extend Decision Tree to Random Froest.
        '''
        self.min_samples_split = min_samples_split
        self.max_depth = max_depth
        self.n_features = n_features
        self.criterion = criterion
        self.root = None # root node

    def fit(self, X, y):
        '''Fit the train dataset by growing a tree'''
        self.n_features = X.shape[1] if not self.n_features else min(self.n_features, X.shape[1])
        self.root = self._grow_tree(X, y)

    def _grow_tree(self, X, y, depth=0):
        '''Recursively building a tree based on greedy searching of the best feature 
        at each step
        '''
        n_samples, n_features = X.shape
        n_labels = len(np.unique(y))
        
        '''Stop crietera. E.g., reach maximum depth, minimum samples at node, 
        node is pure. Then return the most common  value 
        '''
        if depth >= self.max_depth or n_labels == 1 or n_samples < self.min_samples_split:
            leaf_value = self._most_common_label(y)
            return Node(value=leaf_value)
    
        feature_idxs = np.random.choice(n_features, self.n_features, replace=False)
        
        # greedy search for ...
        best_feature, best_threshold = self._best_critera(X, y, feature_idxs)
        left_idxs = np.argwhere(X[:, best_feature] <= best_threshold).flatten()
        right_idxs = np.argwhere(X[:, best_feature] > best_threshold).flatten()
        
        left = self._grow_tree(X[left_idxs, :], y[left_idxs], depth+1)
        right = self._grow_tree(X[right_idxs, :], y[right_idxs], depth+1)
        
        return Node(best_feature, best_threshold, left, right)

    def _most_common_label(self, y):
        '''If multiple labels are found in y, return one of those that are most common'''
        return Counter(y).most_common(1)[0][0]

    def _best_critera(self, X, y, feature_idx):
        '''Find the best feature to maximum information gain
        '''
        best_gain = -1
        split_idx, split_threshold = None, None

        for fi in feature_idx:
            X_column = X[:, fi]
            thresholds = np.unique(X_column)
            for th in thresholds:
                gain = self._information_gain(y, X_column, th)
                if gain > best_gain:
                    best_gain = gain 
                    split_idx = fi 
                    split_threshold = th
        
        return split_idx, split_threshold
    
    def _information_gain(self, y, X_column, threshold):
        '''Calculate the information gain when metric is entropy,
        or Gini information gain when metric is gini
        criterion: {entropy, gini}
        '''
        metric = entropy if self.criterion == 'entropy' else gini

        parent_entropy = metric(y)
        left_idxs = np.argwhere(X_column <= threshold).flatten()
        right_idxs = np.argwhere(X_column > threshold).flatten()
        
        n_l, n_r = len(left_idxs), len(right_idxs)
        if n_l * n_r == 0: return 0
        
        e_l, e_r = metric(y[left_idxs]), metric(y[right_idxs])
        child_entropy = (n_l / len(y)) * e_l + (n_r / len(y)) * e_r
        
        return parent_entropy - child_entropy
        
        
    def predict(self, X):
        '''Traverse the tree to make predictions.'''
        return np.array([self._traverse_tree(x, self.root) for x in X])

        
    def _traverse_tree(self, x, node):
        '''Recursively predict the label of sample x'''
        if node.is_leaf_node(): 
            return node.value
        elif x[node.feature] <= node.threshold:
            return self._traverse_tree(x, node.left)
        else:
            return self._traverse_tree(x, node.right)

        
class RandomForest:
    '''An implementation of Random Forest of decision trees'''
    def __init__(self, n_trees=100, min_samples_split=2, max_depth=100, n_features=None):
        self.n_trees = n_trees
        self.min_samples_split = min_samples_split
        self.max_depth = max_depth
        self.n_features = n_features
        self.trees = []
    
    def fit(self, X, y):
        for _ in range(self.n_trees):
            tree = DecisionTree(min_samples_split=self.min_samples_split, \
                               max_depth=self.max_depth, \
                               n_features=self.n_features)

            tree = DecisionTreeClassifier(max_depth=10)
            X_sample, y_sample = self.bootstrap_sample(X, y)
            tree.fit(X_sample, y_sample)
            self.trees.append(tree)
    
    def bootstrap_sample(self, X, y):
        n_samples = X.shape[0]
        idxs = np.random.choice(n_samples, size=n_samples, replace=True)
        return X[idxs], y[idxs]
        
    def predict(self, X):
        tree_preds = np.array([t.predict(X) for t in self.trees])
        tree_preds = np.swapaxes(tree_preds, 0, 1)
        y_preds = [self._most_common_label(p) for p in tree_preds]
        return np.array(y_preds)

    def _most_common_label(self, y):
        '''If multiple labels are found in y, return one of those that are most common'''
        return Counter(y).most_common(1)[0][0]
  
```

# Experiments
Do experiments on our implementation of Decision Tree

```python
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_breast_cancer, make_classification
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

bc = load_breast_cancer()
X, y = bc.data, bc.target

X, y = make_classification(n_samples=1000, n_features=4,
                            n_informative=2, n_redundant=0,
                            random_state=0, shuffle=True)
X_train, X_val, y_train, y_val = train_test_split(X, y, random_state=0, test_size=0.2)

# =============================================================
# clf = DecisionTree(max_depth=10, criterion='gini')
# clf.fit(X_train, y_train)

# y_preds = clf.predict(X_val)
# print("Naive Decision Tree accuracy score", accuracy_score(y_val, y_preds))
# print("-" * 80)

# =============================================================
dtc = DecisionTreeClassifier(max_depth=10, min_samples_leaf=5)
dtc.fit(X_train, y_train)
y_preds = dtc.predict(X_val)
print("sklearn Decision Tree accuracy score", accuracy_score(y_val, y_preds))
print("-" * 80)

# =============================================================
rf = RandomForest(n_trees=10)
rf.fit(X_train, y_train)
y_preds = rf.predict(X_val)
print("Naive Random Forest accuracy score", accuracy_score(y_val, y_preds))
print("-" * 80)

# =============================================================
rf = RandomForestClassifier(n_estimators=10, min_samples_leaf=5)
rf.fit(X_train, y_train)
y_preds = rf.predict(X_val)
print("sklearn Random Forest accuracy score", accuracy_score(y_val, y_preds))
print("-" * 80)

print("Done")
```
